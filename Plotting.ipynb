```python
import cv2
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "train/"
box_file = "train.csv"
```

```python
images = os.listdir(path)
```

```python
len(images)
```

```python
img = images[1]
```

```python
plt.figure(figsize = (8,10))
img_bgr = cv2.imread(path + img)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
s = img.split(".")[0]
s
```

```python
arr = []
with open("train.csv") as f:
    file = csv.reader(f, delimiter = ",")
    count = 0
    for row in file:
        if count == 0:
            pass
        else:
            if row[0] == s:
                arr.append(row[3])
            else:
                pass
        count += 1
```

```python
arr = list(arr)
arr
```

```python
arr_int = []
for i in arr:
    row = []
    for j in i[1:-1].split(","):
        row.append(int(float(j)))
    arr_int.append(row)
```

```python
arr_int
```

```python
plt.figure(figsize = (10,10))
image = img_rgb.copy()
for i in arr_int:
    x1 = i[0]
    y1 = i[1]
    x2 = x1 + i[2]
    y2 = y1 + i[3]
    image = cv2.rectangle(image, (x1,y1), (x2,y2), (255,0,0), 3)
plt.imshow(image)
```

```python
plt.figure(figsize = (10,10))
img_resized = cv2.resize(img_rgb, (400,400), interpolation = cv2.INTER_AREA)
plt.imshow(img_resized)
```

```python
plt.figure(figsize = (10,10))
image = img_resized.copy()
for i in arr_int:
    x1 = (i[0]*400)//1024
    y1 = (i[1]*400)//1024
    x2 = ((i[0] + i[2])*400)//1024
    y2 = ((i[1] + i[3])*400)//1024
    image = cv2.rectangle(image, (x1,y1), (x2,y2), (255,0,0), 1)
plt.imshow(image)
```

```python
img = 'train_resize/005b0d8bb.jpg'
```

```python
arr = []
with open(path) as f:
    file = csv.reader(f, delimiter = ",")
    count = 0
    for row in file:
        if count == 0:
            pass
        else:
            if row[0] == '005b0d8bb':
                arr.append(row[2:])
            else:
                pass
        count += 1
```

```python
arr
```

```python
arr_int = []
for i in arr:
    row = []
    for j in i:
        row.append(int(float(j)))
    arr_int.append(row)
```

```python
img_bgr = cv2.imread(img)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
for i in arr_int:
    xcen = i[0]
    ycen = i[1]
    w = i[2]
    h = i[3]
    x1 = xcen - (w//2)
    y1 = ycen - (h//2)
    x2 = xcen + (w//2)
    y2 = ycen + (h//2)
    image = cv2.rectangle(img_rgb, (x1,y1), (x2,y2), (255,0,0), 1)
```
