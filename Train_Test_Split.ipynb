```python
import os
from PIL import Image
import shutil
import cv2
import numpy as np
import csv
import random
import pandas as pd
```

```python
def split(path, n):
    with open(path) as f:
        file = csv.reader(f,delimiter=",")
        count = 0
        arr = []
        for row in file:
            if count > 0:
                if row[0] not in arr:
                    arr.append(row[0])
                else:
                    pass
            count += 1
    arr_test = random.sample(arr, n)
    arr_train = []
    for i in arr:
        if i not in arr_test:
            arr_train.append(i)
            
    return arr_train, arr_test
```

```python
path = "training.csv"
arr_train, arr_test = split(path, 200)
```

```python
def split_csv(path, arr):
    with open(path) as f:
        file = csv.reader(f, delimiter=",")
        count = 0
        data = []
        for row in file:
            if row[0] in arr:
                data.append(row)
            else:
                pass
    return data
```

```python
train_data = split_csv("training.csv", arr_train)
```

```python
test_data = split_csv("training.csv", arr_test)
```

```python
data_frame = pd.DataFrame(train_data)
data_frame.to_csv("trainn.csv", header = ["image_id","class","xcen","ycen","width","height"], index=False)
```

```python
data_frame = pd.DataFrame(test_data)
data_frame.to_csv("test.csv", header = ["image_id","class","x_cen","ycen","width","height"], index=False)
```

## Dividing box values by image size

```python
training = "trainn.csv"
testing = "test.csv"
```

```python
def normalize(path):
    arr = []
    count = 0
    with open(path) as f:
        file = csv.reader(f, delimiter=',')
        for row in file:
            if count == 0:
                pass
            else:
                data = row
                values = data[2:]
                value = [float(values[0])/256.0, float(values[1])/256.0, float(values[2])/256.0, float(values[3])/256.0]
                data[2:] = value
                arr.append(data)
            count += 1
    return arr
```

```python
train_arr = normalize(training)
```
