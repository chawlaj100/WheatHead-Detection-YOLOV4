```python
import os
import cv2
import csv
import numpy as np
import pickle
```

```python
arr = []
with open("train.csv") as f:
    file = csv.reader(f, delimiter = ",")
    count = 0
    for row in file:
        if count == 0:
            pass
        else:
            if row[0] not in arr:
                arr.append(row[0])
        count += 1
```

```python
len(arr)
```

```python
arr
```

```python
arr_boxes = []
count = 0
for j in arr:
    with open("train.csv") as f:
        file = csv.reader(f, delimiter = ",")
        arr_box = []
        for row in file:
            if row[0] == j:
                temp = row[3]
                boxes = temp[1:-1].split(",")
                boxes_int = list(map(float, boxes))
                arr_box.append(boxes_int)
        arr_boxes.append(arr_box)
    count += 1
    if count%30 == 0:
        print(count)
```

```python
with open("boxes.txt","wb") as fp:
    pickle.dump(arr_boxes,fp)
```
